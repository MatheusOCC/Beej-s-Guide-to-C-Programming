# Máquinas

## Criando a MLP de Regressão

### Define Classe da Rede MLP de Regressão

```python
#========================================================================================================
#
# Criando a Rede de Classificação
#
#========================================================================================================

class RegressaoMLP(nn.Module):
  def __init__(self, tamanho_entrada, tamanho_oculta, tamanho_saida):
    super(RegressaoMLP, self).__init__()
    self.fc1 = nn.Linear(tamanho_entrada, tamanho_oculta)
    self.fc2 = nn.Linear(tamanho_oculta, tamanho_saida)

  def forward(self, entrada):
    feature = F.relu(self.fc1(entrada))
    saida   = self.fc2(feature)
    return saida

  def treino(self, epochs, entrada_treino, saida_treino, criterio, otimizador):
    perda_arr = []

    for epoch in range(epochs):
      saida_pred = self.forward(entrada_treino)
      perda = criterio(saida_pred, saida_treino)
      perda_arr.append(perda.detach().numpy())
      print('Época %d, Perda: %.4f' % (epoch, perda))
      otimizador.zero_grad()
      perda.backward()
      otimizador.step()

    grafico_perda(perda_arr)

  def teste(self, entrada_teste, saida_teste, criterio):
        with torch.no_grad():
            previsoes = self.forward(entrada_teste)
            mse = mean_squared_error(previsoes, saida_teste)
            r2 = r2_score(saida_teste, previsoes)

        print('mse: '+str(mse))
        print('r2: '+str(r2))
        return mse, r2
```

### Treinando e Avaliando o Modelo

```python
%%time

torch.manual_seed(29)

#========================================================================================================
#
# Hiperparâmetros
#
#========================================================================================================

rmlp_args = {'epoch_num': 250,    # Número de épocas.
             'lr': 5e-3,           # Taxa de aprendizado.
             'weight_decay': 5e-4, # Penalidade L2 (Regularização).
             'hidden_size': 100,
             }

#========================================================================================================
#
# Criando a rede
#
#========================================================================================================

regressao_MLP = RegressaoMLP(tamanho_entrada = len(colunas_entrada_regressao),
                                                   tamanho_oculta = rmlp_args['hidden_size'],
                                                   tamanho_saida = len(colunas_saida_regressao)
                                                   ).to(device)

#========================================================================================================
#
# Define função de perda e otimizador
#
#========================================================================================================

r_criterio = nn.L1Loss().to(device)
r_otimizador = optim.Adam(regressao_MLP.parameters(),
                          lr=rmlp_args['lr'],
                          weight_decay=rmlp_args['weight_decay']
                          )

#========================================================================================================
#
# Treinando o modelo
#
#========================================================================================================

perda_arr_treino = regressao_MLP.treino(rmlp_args['epoch_num'],
                                        t_r_entrada_treino,
                                        t_r_saida_treino,
                                        r_criterio,
                                        r_otimizador,
                                        )

#========================================================================================================
#
# Avaliando o modelo
#
#========================================================================================================

perda_arr_teste = regressao_MLP.teste(t_r_entrada_teste,
                                      t_r_saida_teste,
                                      r_criterio,
                                      )
```

## Criando a MLP de Classificação

### Define Rede MLP de Classificação

```python
#========================================================================================================
#
# Criando a Rede de Classificação
#
#========================================================================================================

class ClassificacaoMLP(nn.Module):
  def __init__(self, tamanho_entrada, tamanho_oculta, tamanho_saida):
    super(ClassificacaoMLP, self).__init__()
    self.fc1 = nn.Linear(tamanho_entrada, tamanho_oculta)
    self.fc2 = nn.Linear(tamanho_oculta, tamanho_saida)

  def forward(self, entrada):
    feature = F.relu(self.fc1(entrada))
    saida   = F.sigmoid(self.fc2(feature))
    return saida

  def treino(self, epochs, entrada_treino, saida_treino, criterio, otimizador):
    perda_arr = []
    for epoch in range(epochs):
      saida_pred = self.forward(entrada_treino)
      perda = criterio(saida_pred, saida_treino)
      perda_arr.append(perda.detach().numpy())
      print('Época %d, Perda: %.4f' % (epoch, perda))
      otimizador.zero_grad()
      perda.backward()
      otimizador.step()

    print('Menor valor: '+ str(min(perda_arr)) + '; na época: ' + str(perda_arr.index(min(perda_arr))))
    return perda_arr

  def teste(self, entrada_teste, saida_teste, criterio):
        with torch.no_grad():
            cmlp_previsoes = self.forward(entrada_teste)
            cmlp_previsoes_binarias = (cmlp_previsoes > 0.5).float().squeeze().cpu().numpy()
            cmlp_saida_teste_np = saida_teste.squeeze().cpu().numpy()

        acuracia = accuracy_score(cmlp_saida_teste_np, cmlp_previsoes_binarias)
        print(f'Acurácia: {acuracia:.4f}')

        matriz_confusao = confusion_matrix(cmlp_saida_teste_np, cmlp_previsoes_binarias)
        print(f'Matriz de Confusão:')
        print(matriz_confusao)

        fpr, tpr, thresholds = roc_curve(cmlp_saida_teste_np, cmlp_previsoes.cpu().numpy())
        roc_auc = auc(fpr, tpr)

        return acuracia, matriz_confusao, fpr, tpr, roc_auc
```

### Treinando e Avaliando o Modelo

```python
%%time

torch.manual_seed(29)

#========================================================================================================
#
# Hiperparâmetros
#
#========================================================================================================

cmlp_args = {'epoch_num': 250,    # Número de épocas.
             'lr': 5e-3,           # Taxa de aprendizado.
             'weight_decay': 5e-4, # Penalidade L2 (Regularização).
             'hidden_size': 100,
             }

#========================================================================================================
#
# Criando a rede
#
#========================================================================================================

classificacao_MLP = ClassificacaoMLP(tamanho_entrada = len(colunas_entrada_classificacao),
                                     tamanho_oculta = cmlp_args['hidden_size'],
                                     tamanho_saida = len(colunas_saida_classificacao)
                                     ).to(device)

#========================================================================================================
#
# Define função de perda e otimizador
#
#========================================================================================================

from torch import optim
c_criterio = nn.BCELoss().to(device)
c_otimizador = optim.Adam(classificacao_MLP.parameters(),
                          lr=cmlp_args['lr'],
                          weight_decay=cmlp_args['weight_decay']
                          )

#========================================================================================================
#
# Treinando o modelo
#
#========================================================================================================

perda_arr_treino = classificacao_MLP.treino(cmlp_args['epoch_num'],
                                            t_c_entrada_treino,
                                            t_c_saida_treino,
                                            c_criterio,
                                            c_otimizador
                                            )
plt.title('Perda VS Época')
plt.xlabel("Perda")
plt.xlabel("Epoca")
plt.plot(np.asarray(perda_arr_treino))
plt.show()

#========================================================================================================
#
# Avaliando o modelo
#
#========================================================================================================

acuracia, matriz_confusao, fpr, tpr, roc_auc = classificacao_MLP.teste(t_c_entrada_teste,
                                                                       t_c_saida_teste,
                                                                       c_criterio,
                                                                       )

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

## Criando a ELM de Regressão

### Define Rede ELM de Regressão

```python
#========================================================================================================
#
# Criando a Rede de Classificação
#
#========================================================================================================

class RegressaoELM():
    def __init__(self, input_size, hidden_size, output_size):
        super(RegressaoELM, self).__init__()
        self.alpha = nn.Parameter(torch.randn(input_size, hidden_size))
        self.beta = nn.Parameter(torch.randn(hidden_size, output_size))
        self.bias = nn.Parameter(torch.zeros(hidden_size))

    def forward(self, x):
        h = (x @ self.alpha + self.bias)
        return h @ self.beta

    def fit(self, x, t):
        # Pseudo-inversa da camada oculta
        temp = x @ self.alpha
        H = (temp + self.bias)
        H_pinv = torch.pinverse(H)
        # Ajustar a camada de saída diretamente
        self.beta = H_pinv @ t

    def evaluate(self, x, t):
        # Predição direta
        y_pred = self.forward(x)
        acc = torch.sum(torch.argmax(y_pred, dim=1) == torch.argmax(t, dim=1)).item() / len(t)
        return acc
```
